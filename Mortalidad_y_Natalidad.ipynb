# Mortalidad

```python
url_mort = "http://www.deis.cl/wp-content/uploads/2016/08/3.-Defunciones-seg%C3%BAn-grupo-de-edad-regi%C3%B3n-gran-grupo-de-causas-y-sexo.-Chile-2014.xlsx"

df_mort = pd.read_excel(url_mort, skiprows=4, skipfooter=16)
display(df_mort.head())



print(df_mort.columns)

df_mort_main = df_mort[df_mort['Unnamed: 0'].notnull()].copy()
df_mort_main.rename(columns={'Unnamed: 0':'Region_CIE10','Unnamed: 1':'Enfermedad',
                             'Unnamed: 3':'Total','Unnamed: 21':'Ignorado'}, inplace = True)
display(df_mort_main.head())

def hasNumbers(inputString):
    return any(char.isdigit() for char in inputString)

df_mort_main['Region_CIE10']

df_mort_main['Region_CIE10'].apply(hasNumbers)

df_mort_region = df_mort_main[~df_mort_main['Region_CIE10'].apply(hasNumbers)][1:].set_index('Region_CIE10')

(100*df_mort_region.drop(['Enfermedad','Unnamed: 2','Total','Ignorado'],1)).head(5)
#.div(df_mort_region.Total.astype(np.float),0).info()#.T.plot()
```
