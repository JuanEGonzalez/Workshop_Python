### Definamos Función que haga lo anterior

```python
# Así habíamos decidido la Web:
id_i = '18'
date_ini = '180522'
date_end = '180621'
frecuencia = 'diario'#horario

url = ("https://sinca.mma.gob.cl/cgi-bin/APUB-MMA/apub.tsindico2.cgi?outtype=xcl&macro=./RM/D"+id_i+
       "/Cal/PM10//PM10."+frecuencia+"."+frecuencia+".ic&from="+date_ini+"&to="+date_end+
       "&path=/usr/airviro/data/CONAMA/&lang=esp&rsrc=&macropath=")
# Y después leímos y realizamos transformaciones.
```

```python
# Usamos lo aprendido recién para definir una función que lea la info que yo necesito, la transforme y me la entregue lista
def read_csv_M10(id_i,date_ini,date_end,frecuencia):
    url = ("https://sinca.mma.gob.cl/cgi-bin/APUB-MMA/apub.tsindico2.cgi?outtype=xcl&macro=./RM/D"+id_i+
           "/Cal/PM10//PM10."+frecuencia+"."+frecuencia+".ic&from="+date_ini+"&to="+date_end+
           "&path=/usr/airviro/data/CONAMA/&lang=esp&rsrc=&macropath=")
    df = pd.read_csv(url, sep = ';')
    df.rename(columns={'Registros preliminares':'MP10'}, inplace = True)    
    df['estacion_id'] = np.int(id_i)# Nuevo para que guarde también la información del id de la estación
    if (frecuencia == 'diario'):
        df['FECHA'] =pd.to_datetime(df['FECHA (YYMMDD)'], format = "%y%m%d")
        return df[['estacion_id','FECHA','MP10']]# Así como puedo seleccionar columnas puedo reordenar!
    if (frecuencia == 'horario'):
        df['HORA (HHMM)'] = df['HORA (HHMM)'].astype(np.str)
        df.loc[df['HORA (HHMM)'].str.len()  == 3, 'HORA (HHMM)'] = '0' + df.loc[df['HORA (HHMM)'].str.len() < 100, 'HORA (HHMM)'] 
        df.loc[df['HORA (HHMM)'].str.len()  == 1, 'HORA (HHMM)'] = '0000'
        df['FECHA_HORA_RAW'] = df['FECHA (YYMMDD)'].astype(np.str) + '-' + df['HORA (HHMM)']
        df['FECHA_HORA'] =pd.to_datetime(df['FECHA_HORA_RAW'], format = "%y%m%d-%H%M")
        return df[['estacion_id','FECHA_HORA','MP10']] 
```

Es una herramienta muy útil para combinar 2 DataFrames con elementos similares

Pandas: rolling. Util para aplicar una función en una ventana de tiempo.

```python
df_11 = df_estaciones_M10[df_estaciones_M10.estacion_id == 11].set_index('FECHA_HORA')
df_2show = df_11.MP10.fillna(method = 'ffill').rolling(window = 24, center = True).mean()
df_2show.sample(5)
```

## Ejemplo 4. Mapa!

```python
# Funcion que convierte LAT y LONG en la proyección Mercator (plano 2D)
import math
from ast import literal_eval
def merc(Lat, Lon):#Coords):
    #Coordinates = literal_eval(Coords)
    #lat = Coordinates[0]
    #lon = Coordinates[1]
    
    r_major = 6378137.000
    x = r_major * math.radians(Lon)
    scale = x/Lon
    y = 180.0/math.pi * math.log(math.tan(math.pi/4.0 + 
        Lat * (math.pi/180.0)/2.0)) * scale
    return (x, y)
```

```python
df_estaciones['coords_x'] = df_estaciones.apply(lambda x: merc(x['lat'], x['lon'])[0], axis = 1)
df_estaciones['coords_y'] = df_estaciones.apply(lambda x: merc(x['lat'], x['lon'])[1], axis = 1)
df_estaciones
```
